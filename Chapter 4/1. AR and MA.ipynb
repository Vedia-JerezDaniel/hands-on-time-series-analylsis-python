```python
import pandas as pd 
```

```python
df = pd.read_csv(r'..\Data\daily-total-female-births-CA.csv')
```

```python
df['births'].head()
```

```python
import statsmodels as sm
import matplotlib.pyplot as plt
model = sm.tsa.arima.model.ARIMA(df['births'], order=(1, 1, 0))  
results_AR = model.fit()  
plt.plot(results_AR.fittedvalues, color='black')
plt.ylabel('Autoregressive (1)')
plt.xlabel('Time')
plt.rcParams["figure.figsize"] = [9,6]
plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
model = ARIMA(df['births'], order=(2, 1, 0))  
results_AR = model.fit()  
plt.plot(results_AR.fittedvalues, color='black')
plt.ylabel('Autoregressive (2)')
plt.xlabel('Time')
plt.rcParams["figure.figsize"] = [9,6]
plt.show()
```

```python
import matplotlib.pyplot as plt
model = ARIMA(df['births'], order=(1, 1, 0))  
results_AR = model.fit()  
plt.plot(results_AR.fittedvalues, color='black')
plt.ylabel('Moving Average(1)')
plt.xlabel('Time')
plt.rcParams["figure.figsize"] = [16,9]
plt.show()
```
